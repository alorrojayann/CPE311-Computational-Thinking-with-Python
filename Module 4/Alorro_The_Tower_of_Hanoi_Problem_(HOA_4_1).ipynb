## **Hands-on Activity 4.1: The Tower of Hanoi Problem**

Rules of the problem:
- Only one disk can be transfer at a time.
- Each move consists of taking the upper disk from one of the peg and placing it on the top of another peg i.e. a disk can only be moved if it is the uppermost disk of the peg.
- Never a larger disk is placed on a smaller disk during the transfer.

Tower-of-Hanoi-Tower-of-Brahma-or-Lucas-Tower.png

```python
def toh(n, source, auxiliary, destination):
  # base case
  if n == 1:
    print('Disk', n, 'FROM:', source, '-->', destination)
  else:
    toh(n-1, source, destination, auxiliary)
    print('Disk', n, 'FROM:', source, '-->', destination)
    toh(n-1, auxiliary, source, destination)
```

**Breakdown of Code**



*   The funtion 'toh' is defines which take 4 parameters:
    *   n = no. of disks
    - source = name of the starting peg
    - auxiliary = temporary storage peg
    - destination = the peg in which the disks needs to be moved



*   Base case = 1. Prints a message that the single disk in the stack is simply moved to the destination peg, solving the problem.
*   Recusive Implementation (when n > 1):
    - (recursive call) move 'n-1' from source to auxiliary, the destination is used as the temporary storage in this step
    - move the disks left to the destination peg
    - (recursive call) move 'n-1' disks of auxiliary to destination, the source is used as the temporary storage in this step
- The recusion stops until the base case is met. The solution/steps to the problem is then displayed.







```python
# example use
n = 3
toh(n, 'Source', 'Auxiliary', 'Destination')
```
